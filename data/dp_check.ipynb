```python
from bs4 import BeautifulSoup

def extract_text_from_element(content):
    """
    content: {
        "html": "...",
        "markdown": "...",
        "text": "..."
    }
    Extracts and combines visible text from html + markdown + OCR text.
    """
    html = content.get("html", "")
    soup = BeautifulSoup(html, "html.parser")

    # <br> → 줄바꿈
    for br in soup.find_all("br"):
        br.replace_with("\n")

    # HTML에서 추출한 텍스트
    parts = [soup.get_text(separator="\n").strip()]

    # markdown / text도 있으면 이어붙이기
    markdown = content.get("markdown")
    if markdown:
        parts.append(markdown.strip())

    text = content.get("text")
    if text:
        parts.append(text.strip())

    # 줄바꿈으로 구분해서 합치기
    return "\n".join([p for p in parts if p])
```

```python
def chunk_by_heading1(elements):
    chunks = []
    current_chunk = None

    for elem in elements:
        category = elem.get("category")
        content = elem.get("content", {})
        text = extract_text_from_element(content)

        if category == "heading1":
            if current_chunk:
                chunks.append(current_chunk)

            current_chunk = {
                "section_heading": text,
                "elements": []
            }
        else:
            if current_chunk:
                current_chunk["elements"].append({
                    "category": category,
                    "text": text,
                    "id": elem.get("id"),
                    "page": elem.get("page")
                })

    if current_chunk:
        chunks.append(current_chunk)

    return chunks
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("nlpai-lab/KURE-v1")  # 한글 문서에 적합한 모델
```

```python
import chromadb
from chromadb.config import Settings

client = chromadb.PersistentClient(path="./chroma_db")  # 로컬 디스크 저장
collection = client.get_or_create_collection("lg_manual")
```

```python
chunks
```

```python
documents = []
metadatas = []
ids = []

for i, chunk in enumerate(chunks):  # chunks는 너가 만든 위 리스트
    section = chunk["section_heading"]
    for elem in chunk["elements"]:
        text = elem["text"]
        if not text.strip():
            continue

        doc_id = f"LG_{i}_{elem['id']}"
        documents.append(text)
        ids.append(doc_id)
        metadatas.append({
            "section": section,
            "category": elem["category"],
            "page": elem["page"],
            "element_id": elem["id"]
        })
```

```python
metadatas
```

```python
embeddings = model.encode(documents).tolist()

collection.add(
    documents=documents,
    embeddings=embeddings,
    metadatas=metadatas,
    ids=ids
)
```

```python
query_embedding = model.encode(["일반적인 청소할 땐 어떻게 해야하나요?"])
query_embedding.shape
```

```python

results = collection.query(
    query_embeddings=[query_embedding],
    n_results=5
)

for doc, meta in zip(results['documents'][0], results['metadatas'][0]):
    print("Text:", doc)
    print("Meta:", meta)
    print("---")
```

```python
from bs4 import BeautifulSoup
import json

html_content = response.json()["content"]["html"]

soup = BeautifulSoup(html_content, "html.parser")

chunks = []
current_chunk = {"section": None, "elements": []}

for tag in soup.find_all():
    tag_data = {
        "tag": tag.name,
        "text": tag.get_text(separator="\n").strip(),
        "id": tag.get("id"),
        "data_category": tag.get("data-category"),
        "font_size": tag.get("style"),
    }
    if tag.name == "h1":
        if current_chunk["elements"]:
            chunks.append(current_chunk)
            current_chunk = {"section": tag.get_text().strip(), "elements": []}
        else:
            current_chunk["section"] = tag.get_text().strip()
    else:
        current_chunk["elements"].append(tag_data)

if current_chunk["elements"]:
    chunks.append(current_chunk)

import pandas as pd

df_chunks = pd.DataFrame([
    {
        "section": chunk["section"],
        "content": "\n".join([el["text"] for el in chunk["elements"]])
    }
    for chunk in chunks
])
```
